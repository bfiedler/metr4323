# The Collatz Conjecture

v1.0 25 April 2018

Fun with the [Collatz Conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture).  Let's use the generation of this sequence $a_i$ for $i=0,1,2,3\ldots$ as a very simple analog to making a weather forecast.  Note the wikipedia article reserves $n$ to be the initial value $a_0$. In this Python code, we update a variable `n`.


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generates the sequence
def collatz(n):
    assert type(n) == type(3) and n>=1 # avoid infinite loops, start with positive integer
    sequence = [n]
    while n != 1:
        if n%2 == 0:
            n = n//2  # integer divide
        else:
            n = 3*n +1
        sequence.append(n)
    return sequence
```

```python
collatz(3)
```

```python
for n in range(1,15):
    print( collatz(n) )
```

```python
#a quick plot
n = 27
plt.plot(collatz(n));
```

```python
# a fancier plot
n=27 # try 27 or other positive integers
myfig, myaxes = plt.subplots(facecolor='lightgrey',figsize=(10,4))
myaxes.set_axis_bgcolor('yellow')
myaxes.plot(collatz(n))
myaxes.set_xlabel(r'$i$',fontsize=30)
myaxes.set_ylabel(r"$a_i$",fontsize=30,rotation=0)
myaxes.set_title('Collatz sequence initialized with n=' + str(n), fontsize=30 );
```

```python
# study only the length of the Collatz sequence
isteps=[]
for n in range(1,10000): 
    istep = len(collatz(n)) 
    isteps.append(istep)
# Note: the Wikipedia defines "total stopping time" to be len(collatz(n))-1
```

```python
# see Wikipedia for a similar plot
myfig2, myaxes2 = plt.subplots(facecolor='lightgrey',figsize=(8,6))
myaxes2.set_axis_bgcolor('white')
myaxes2.plot(isteps,'r.')
myaxes2.set_xlabel(r'$n$',fontsize=30)
myaxes2.set_title('length of Collatz sequence initialized with n', fontsize=18 );
```
